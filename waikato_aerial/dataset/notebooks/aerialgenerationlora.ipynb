```python
import bson
from huggingface_hub.repocard import RepoCard
from diffusers import StableDiffusionPipeline
import torch


pipe = StableDiffusionPipeline.from_pretrained("stabilityai/stable-diffusion-2-1-base", torch_dtype=torch.float16).to("cuda")

def generate(
    token: str,
    prompts: list[str], 
    images_per_prompt: int = 20, 
    num_inference_steps: int = 25
) -> None:
    class_name = token.lower()
    lora_model_id = f"dushj98/{class_name}_lora"
    pipe.load_lora_weights(lora_model_id)
    
    generation_dir = f"generated/{class_name}"
    for p in prompts:
        for i in range(images_per_prompt):
            image = pipe(p, num_inference_steps=num_inference_steps).images[0]
            image.save(f"generated/{class_name}/{str(bson.ObjectId())}.png")
    
    print(f"✅ {token}\n")
```

```python
import os
import shutil
from datasets import load_dataset


def push_to_hub(token: str):
    class_name = token.lower()
    os.makedirs(f"hfd/{class_name}", exist_ok=True)
    
    for idx, f in enumerate(sorted([i for i in os.listdir(f"generated/{class_name}") if i.endswith(".png")])):
        shutil.copy(f"generated/{class_name}/{f}", f"hfd/{class_name}/{idx}.png")

    dataset = load_dataset("imagefolder", name=f"aerial_synthetic_{class_name}", data_dir=f"hfd/{class_name}")
    dataset.push_to_hub(f"dushj98/aerial_synthetic_{class_name}", token=HFT)
    return dataset
    
```
