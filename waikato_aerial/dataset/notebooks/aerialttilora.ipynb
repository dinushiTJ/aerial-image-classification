## Data Prep

```python
classes = {
    'broadleaved_indigenous_hardwood': 'BIH', 
    'deciduous_hardwood': 'DHW', 
    'grose_broom': 'GBM', 
    'harvested_forest': 'HFT', 
    'herbaceous_freshwater_vege': 'HFV', 
    'high_producing_grassland': 'HPG', 
    'indigenous_forest': 'IFT', 
    'lake_pond': 'LPD', 
    'low_producing_grassland': 'LPG', 
    'manuka_kanuka': 'MKA', 
    'shortrotation_cropland': 'SRC', 
    'urban_build_up': 'UBU', 
    'urban_parkland': 'UPL'
}
```

```python
!rm -r dataset
!rm -r lora
!rm -r generated

import os

for v in classes.values():
    os.makedirs(f"dataset/{v.lower()}", exist_ok=True)
    os.makedirs(f"tti_lora/{v.lower()}_tti_lora", exist_ok=True)
```

```python
import os
from datasets import load_dataset

dataset = load_dataset("dushj98/waikato_aerial_2017_sd_ft", split="train")
for i, r in enumerate(dataset):
    image = r["image"]
    image.save(os.path.join("dataset", r["token"].lower(), f"{i}.png"))
```
